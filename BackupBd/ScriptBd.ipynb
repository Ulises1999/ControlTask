# [dbo].[UpdateTask]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateTask]    Script Date: 17/10/2022 16:01:57 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- Store Procedure=====================================================================
CREATE PROCEDURE [dbo].[UpdateTask] (@id int, @name varchar(50), @note varchar(50))
AS BEGIN
-- Actualizar Titulo y Notas
UPDATE dbo.Task
SET Name = @name , Note = @note
WHERE Taskid = @id
END
GO
```
